```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(x=df['cgpa'],y=df['iq'],hue=df['placement'])
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
p.fit(X,y)
```

```python
p.coef_
```

```python
p.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X=X.values,y=y.values,clf=p,legend=2)
```
